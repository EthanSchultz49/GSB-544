```python
def recommend_trails(data,national_park, min_elev=None, max_elev=None, min_rating=None,type=None):

    if national_park is None:
        national_park = data[data["name"]== national_park]
    if min_elev is None:
        min_elev = data[data["elevation_gain"] >= min_elev]
    if max_elev is None:
        max_elev = data[data["elevation_gain"] <= max_elev]
    if min_rating is None:
        min_rating = data[data["difficulty_rating"] <= min_rating]
    if type is None:
        type = data[data["route_type"] == type]    
    return data["name"]
```
