```python
import pandas as pd
```

```python
import requests
response = requests.get("https://worldpopulationreview.com/state-rankings/starbucks-stores-by-state#google_vignette")
```

```python
from bs4 import BeautifulSoup
soup = BeautifulSoup(response.content, "html.parser")
```

```python
table = soup.find("table", class_="wpr-table")
table
```

```python
#1
# initialize an empty list
rows = []

# iterate over all rows in the faculty table
for coffee in table.find_all("tr")[1:]:
    state = coffee.find_all("th")
    cells = coffee.find_all("td")

    # get the state names
    state_tag = state[0].find("a") or state[0]
    states = state_tag.get_text(strip = True)

    # get the different data for years
    year2023_tag = cells[0].find("td") or cells[0]
    year2023 = year2023_tag.get_text(strip = True)

    year2024_tag = cells[2].find("td") or cells[2]
    year2024 = year2024_tag.get_text(strip = True)

    year2021_tag = cells[1].find("td") or cells[1]
    year2021 = year2021_tag.get_text(strip = True)

    # Append this data.
    rows.append({
        "States": states, 
        "Starbucks 2021": year2021,
        "Starbucks 2023": year2023,
        "Starbucks 2024": year2024
       
    })
```

```python
starbucks = pd.DataFrame(rows)
starbucks
```

```python
response = requests.get("https://worldpopulationreview.com/state-rankings/dunkin-donuts-by-state")
soup = BeautifulSoup(response.content, "html.parser")
table1 = soup.find("table", class_="wpr-table")
```

```python
#1
# initialize an empty list
rows1 = []

# iterate over all rows in the faculty table
for coffee in table1.find_all("tr")[1:]:
    state = coffee.find_all("th")
    cells = coffee.find_all("td")

    # get the state names
    state_tag = state[0].find("a") or state[0]
    states = state_tag.get_text(strip = True)

    # get the different data for years
    year2024_tag = cells[0].find("td") or cells[0]
    year2024 = year2024_tag.get_text(strip = True)

    year2023_tag = cells[1].find("td") or cells[1]
    year2023 = year2023_tag.get_text(strip = True)

    # Append this data.
    rows1.append({
        "States": states, 
        "Dunkin-Donuts 2023": year2023,
        "Dunkin-Donuts 2024": year2024,
       
    })
```

```python
dunkin = pd.DataFrame(rows1)
dunkin
```

```python
merged = starbucks.merge(dunkin, on = ["States"], 
how = "outer")
merged.head()
long = merged.melt(id_vars=["States"], value_vars=["Starbucks 2021", "Starbucks 2023", "Starbucks 2024", "Dunkin-Donuts 2023", "Dunkin-Donuts 2024"], var_name="Year_Brand", value_name="stores")
# Split the 'Year_Brand' column into 'Year' and 'Brand'
long[['Brand', 'Year']] = long['Year_Brand'].str.split(' ', expand=True)
# Drop the 'Year_Brand' column if no longer needed
long = long.drop(columns=['Year_Brand'])
```

```python
response = requests.get("https://simple.wikipedia.org/wiki/List_of_U.S._states_by_population")
soup = BeautifulSoup(response.content, "html.parser")

tables = soup.find_all("table", class_ = "wikitable")

#get first table that we want
table = tables[0]

rows = []

for tr in table.find_all("tr")[1:]:
    cells = tr.find_all("td")

    # get states
    state_tag = cells[2].find("a") or cells[2]
    state = state_tag.get_text(strip = True)

    # get population
    pop_tag = cells[3].find("a") or cells[3]
    population = pop_tag.get_text(strip = True)

    # append data
    rows.append({
        "States" : state,
        "population" : population
    })

# convert to a df
population = pd.DataFrame(rows)
population.head()

withpop = long.merge(population, on = "States", how = "left")
withpop
```

```python
# Automate
import re
def scrape_store_data(url):
    response = requests.get(url)
    soup = BeautifulSoup(response.content, "html.parser")
    
    table = soup.find("table")
    
    company_name = re.search(r'state-rankings/(.+?)-stores', url)
    company_name = company_name.group(1).replace("-", " ").title() if company_name else "Unknown"
    
    rows = []

    for coffee in table.find_all("tr")[1:]:
        state = coffee.find_all("th")
        cells = coffee.find_all("td")

        state_tag = state[0].find("a") or state[0]
        states = state_tag.get_text(strip=True)

        year2023 = cells[0].get_text(strip=True)
        year2021 = cells[1].get_text(strip=True)
        year2024 = cells[2].get_text(strip=True)

        rows.append({
            "States": states,
            f"{company_name} 2021": year2021,
            f"{company_name} 2023": year2023,
            f"{company_name} 2024": year2024
        })

    # Convert to DataFrame
    df = pd.DataFrame(rows)
    
    return df

# Example usage
url = "https://worldpopulationreview.com/state-rankings/walmart-stores-by-state"
df = scrape_store_data(url)
print(df)
```
