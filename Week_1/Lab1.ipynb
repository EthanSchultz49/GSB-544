#1
 It seems that we have a bubble plot here which maps out the income and life expectacy of different world regions. There is a fill attribute which allows up to see how the world regions differ in respect to their income and life expectancies. Also it seems there is a grouping aesthetics for the x variable of income which allows you to see it in four differernt levels. Along with the world regions there is another aesthetic which allows you to see the relative population size of each place within a world region based off the size of the bubble for it.

```python
#2
import pandas as pd
data1 = pd.read_csv("/Users/ethanschultz/Documents/GSB 544/Data Files/q1data.csv")
```

```python
import plotnine as pl
(pl.ggplot(data1, pl.aes(x='income', y='life_exp', size='population', fill='four_regions'))
     + pl.geom_point(alpha=0.9)
     + pl.scale_size(range=(2, 20))
     + pl.annotate('text', x=5000, y=65, label=str('2010'), size=100, alpha =.2, color="gray")
     + pl.labs(x="Income", y="Life Expectancy")
     + pl.theme(axis_title=pl.element_text(size=12), title=pl.element_text(size=14), panel_background= pl.element_text(color="White"), 
       panel_grid_major = pl.element_line(color="gray", alpha = 0.1))
     + pl.scale_x_log10(breaks = [500,1000,2000,4000,8000,16000,32000,64000],
     labels = ['500','1000','2000','4000','8000','16k','32k','64k'], limits = [500,100000])
     + pl.scale_y_continuous(breaks = [20,30,40,50,60,70,80,90])
    )
```

```python
(pl.ggplot(data1, pl.aes(x='income', y='life_exp', size='population', fill='four_regions'))
     + pl.geom_polygon(alpha=0.9)
     + pl.scale_size(range=(2, 20))
     + pl.annotate('text', x=30000, y=65, label=str('2010'), size=100, alpha =.2, color="gray")
     + pl.labs(x="Income", y="Life Expectancy")
     + pl.theme(axis_title=pl.element_text(size=12), title=pl.element_text(size=14), panel_background= pl.element_text(color="White"), 
       panel_grid_major = pl.element_line(color="gray", alpha = 0.1))
     + pl.xlim(0,65000)
     + pl.scale_y_log10(breaks = [20,30,40,50,60,70,80,90])
    )
```

```python
data2 = pd.read_csv("/Users/ethanschultz/Documents/GSB 544/Data Files/q2data.csv")
data2['four_regions'] = pd.Categorical(data2['four_regions'])
data2['exports'] = pd.to_numeric(data2['exports'], errors='coerce')
data2['imports'] = pd.to_numeric(data2['imports'], errors='coerce')
data2['energy'] = pd.to_numeric(data2['energy'], errors='coerce')


(pl.ggplot(data2, pl.aes(x='exports', y='imports', size='energy', fill='four_regions'))
     + pl.geom_point(alpha=0.9)
     + pl.scale_size(range=(2, 20))
     + pl.annotate('text', x=100, y=200, label=str('1997'), size=100, alpha =.2, color="gray")
     + pl.labs(x="Exports (\% of GDP)", y="Imports (\% of GDP)")
     + pl.theme(axis_title=pl.element_text(size=12), title=pl.element_text(size=14), panel_background= pl.element_text(color="White"), 
       panel_grid_major = pl.element_line(color="gray", alpha = 0.1))
     + pl.scale_x_continuous(limits=(0, 220), breaks=[20, 40, 60, 80, 100, 120, 140, 160, 180, 200, 220], expand=(0, 0))
     + pl.scale_y_continuous(limits=(0, 400), breaks=[50, 100, 150, 200, 250, 300, 350, 400], expand=(0, 0))
    )
```

```python
(pl.ggplot(data2, pl.aes(x='exports', y='imports', size='energy', color='four_regions'))
     + pl.geom_line(alpha=0.9)
     + pl.scale_size(range=(2, 20))
     + pl.annotate('text', x=100, y=200, label=str('1997'), size=100, alpha =.2, color="gray")
     + pl.labs(x="Exports (\% of GDP)", y="Imports (\% of GDP)")
     + pl.theme(axis_title=pl.element_text(size=12), title=pl.element_text(size=14), panel_background= pl.element_text(color="White"), 
       panel_grid_major = pl.element_line(color="gray", alpha = 0.1))
     + pl.scale_x_continuous(limits=(0, 220), breaks=[20, 40, 60, 80, 100, 120, 140, 160, 180, 200, 220], expand=(0, 0))
     + pl.scale_y_continuous(limits=(0, 400), breaks=[50, 100, 150, 200, 250, 300, 350, 400], expand=(0, 0))
    )
```
