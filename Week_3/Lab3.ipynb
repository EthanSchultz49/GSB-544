```python
def pluralize_gift(gift):
  """
  Returns plural of a noun
  
  Parameters
  ----------
  gift: str
    A noun
    
  Return
  ------
  str
    Plural version
  """

  if "oo" in gift:
    gift = gift.replace("oo","ee")
  elif gift.endswith("y"):
    gift = gift[:-1] + "ies"
  else:
    gift = gift + "s"

  return gift
```
